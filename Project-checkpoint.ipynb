# task 3: 

```python
# pandas date time
data['Order Date'] = pd.to_datetime(data['Order Date'])
data_d1 = data.set_index('Order Date')
data_d1
```

```python
# pandas date time
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
data_d2 = data.set_index('Ship Date')
data_d2
```

```python
# add price per unit and add column
data["Total Price"] = data["Sales"] / data["Quantity"]
cols=list(data.columns)
data = data[cols[0:17]+[cols[-1]]+cols[17:22]]
data
```

```python
data["Cost"] = data["Sales"] + data["Profit"]
data
```

```python
#Grouping the data month wise
data_df=pd.DataFrame()
data_df['Month_year'] = pd.to_datetime(data['Order Date']).dt.to_period('M')
data_df['Cost1'] = data['Cost']
data_df['Profit1'] = data['Profit']
group1=data_df.groupby(['Month_year'])

data_df_month=pd.DataFrame()
data_df_month = group1['Cost1'].agg(np.sum).to_frame(name='Monthly_revenue').reset_index()
data_df_month = data_df_month.astype({'Month_year':str,'Monthly_revenue':int})
print(data_df_month)
```

```python
# plot of the graph
from matplotlib import style
```

```python
style.use('ggplot')
plt.figure(figsize=(50,10))
plt.bar(data_df_month['Month_year'],data_df_month['Monthly_revenue'])
plt.xlabel('Month_Year')
plt.ylabel('Sale')
plt.show()
```

```python
data_df_month.describe()
```

```python
# Find out the most and least sold product category from the store
#quantity based
data_df10 = pd.DataFrame()
data_df10['Category'] = data['Category']
data_df10['Quantity_1'] = data['Quantity']
group7 = data_df10.groupby(['Category'])
data_df10 = group7['Quantity_1'].agg(np.sum).to_frame(name='Total').reset_index()
data_df10 = data_df10.astype({'Category':str,'Total':int})
print(data_df10)
data['Category'].value_counts().plot.bar()
plt.xlabel('Category')
plt.ylabel('np Category')
plt.show()
```
